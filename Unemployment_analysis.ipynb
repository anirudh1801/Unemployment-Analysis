```python
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = ["States", "Date", "Frequency",
              "Estimated Unemployment Rate", "Estimated Employed",
              "Estimated Labour Participation Rate", "Region",
              "longitude", "latitude"]

plt.title("Indian Unemployment")
sns.histplot(data=df, x="Estimated Employed", hue="Region")
plt.show()

```

```python
#letâ€™s see the unemployment rate according to different regions of India:

plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(data=df, x="Estimated Unemployment Rate", hue="Region")
plt.show()
```

```python
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
#Now letâ€™s create a dashboard to analyze the unemployment rate of each Indian state by region. For this, weâ€™ll use a sunburst plot:
import plotly.express as px
import plotly.io as pio

# Define the DataFrame subset and the chart
unemployment = df[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(
    unemployment,
    path=["Region", "States"],
    values="Estimated Unemployment Rate",
    width=700,
    height=700,
    color_continuous_scale="RdY1Gn",
    title="Unemployment Rate in India"
)

# Open the chart directly in the browser
pio.show(figure, renderer="browser")
```
